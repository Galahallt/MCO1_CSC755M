In this project, we will discuss the implementation and analysis (with frequency counts) of pi approximation algorithms. Here, we will talk about 3 variants of pi approximation algorithms and see if the value of pi is better computed as input size increases.


Let us first initialize the libraries to be used to better analyze the pi approximation algorithms.

```python
import random
import time 
import matplotlib.pyplot as plt
```

Let us also initialize the arrays that we will need to visualize the growth rate of the inputs over time.

```python
var1_time = []

var2_time = []

var3_time = []
```

For all the algorithms to be presented, we compute pi by only considering one quadrant since the approximation of pi is just computed as 4 times the proportion of the dots inside and outside the circle. 

### Variant 1 algorithm

The first algorithm computes the approximation of pi by only considering the integer values of x and y up to r. The next cell will show us the code for the first algorithm.

```python
def var1(r):
  start = time.time()
  inCount = 0
  rsquare = r*r
  
  for x in range(1, r+1):
    xsquare = x*x
    for y in range(1, r+1):
      if (xsquare + y*y <= rsquare):
        inCount+=1
  print("--------------------Var1--------------------")
  print("input size: ", rsquare)
  print("pi value: ", (4*inCount/rsquare))
  end = time.time() - start
  print("Time elapsed: %s seconds" % end)
  return rsquare, end
```

Let us see how fast and accurate would the first algorithm approximate the value of pi given different input size.

```python
# Append the results to array to be plotted later
var1_time.append(var1(100))
var1_time.append(var1(250))
var1_time.append(var1(500))
var1_time.append(var1(750))
var1_time.append(var1(1000))
var1_time.append(var1(2500))
var1_time.append(var1(5000))
var1_time.append(var1(7500))
var1_time.append(var1(10000))
```

The function below is the code for plotting the data points of the first algorithm

```python
def var1_plot(r):
  start = time.time()
  x_inside = []
  y_inside = []
  x_outside = []
  y_outside = []
  inCount = 0
  rsquare = r*r

  for x in range(1, r+1):
    xsquare = x*x
    for y in range(1, r+1):
      if (xsquare + y*y <= rsquare):
        x_inside.append(x)
        y_inside.append(y)
        inCount+=1
      else:
        x_outside.append(x)
        y_outside.append(y)
  print("--------------------Var1 w/ Visualization--------------------")
  print("input size: ", rsquare)
  print("pi value: ", (4*inCount/rsquare))
  print("Time elapsed: %s seconds" % (time.time() - start))
  if (rsquare <= 1000000):
    fig, ax = plt.subplots(1)
    ax.set_aspect('equal')
    ax.scatter(x_inside, y_inside, c='b', alpha=0.8, edgecolors=None) # points inside the circle will be colored blue 'b'
    ax.scatter(x_outside, y_outside, c='r', alpha=0.8, edgecolors=None) # points outside the circle will be colored red 'r'
```

```python
var1_plot(10)
var1_plot(50)
var1_plot(100)
var1_plot(1000)
```

Here is the execution time of variant 1 algorithm as input size increases:

```python
var1_timeplot = list(zip(*var1_time))
plt.xlabel('Time (in seconds)')
plt.ylabel('Input Size')
plt.plot(var1_timeplot[1], var1_timeplot[0])
```

The first algorithm actually has the time complexity of O(r^2) rather than O(n^2) since we used a proxy for the size of the input which is r. The function also utilized nested loops with the range of r which resulted to O(r^2) time complexity.

### Variant 2 algorithm

The second algorithm computes the approximation of pi by randomly generating N (input size) x-y pairs. These x-y pairs are real number values in the range of 0 to 1. The next cell will show us the code for the second algorithm.

```python
def var2(num_points):
  start = time.time()
  inCount = 0
  outCount = 0

  for i in range(num_points):
    x = random.random()
    y = random.random()
    if (x*x + y*y <= 1):
      inCount+=1
    else:
      outCount+=1
  print("--------------------Var2--------------------")
  print("input size: ", num_points)
  print("pi value: ", (4*inCount/(inCount+outCount)))
  end = time.time() - start
  print("Time elapsed: %s seconds" % end)
  return num_points, end
```

Let us see the approximation of pi if we passed the same number as what we did in the first algorithm.

```python
# Append the results to array to be plotted later
var2(100)
var2(250)
var2(500)
var2(750)
var2(1000)
var2(2500)
var2(5000)
var2(7500)
var2(10000)
```

Let us see how fast and accurate would the first algorithm approximate the value of pi given the same input sizes used in the first algorithm.

```python
var2_time.append(var2(10000))
var2_time.append(var2(62500))
var2_time.append(var2(250000))
var2_time.append(var2(562500))
var2_time.append(var2(1000000))
var2_time.append(var2(6250000))
var2_time.append(var2(25000000))
var2_time.append(var2(56250000))
var2_time.append(var2(100000000))
```

The function below is the code for plotting the data points of the second algorithm

```python
def var2_plot(num_points):
  start = time.time()
  x_inside = []
  y_inside = []
  x_outside = []
  y_outside = []
  inCount = 0
  outCount = 0

  for i in range(num_points):
    x = random.random()
    y = random.random()
    if (x*x + y*y <= 1):
      x_inside.append(x)
      y_inside.append(y)
      inCount+=1
    else:
      x_outside.append(x)
      y_outside.append(y)
      outCount+=1
  print("--------------------Var2 w/ Visualization--------------------")
  print("input size: ", num_points)
  print("pi value: ", (4*inCount/(inCount+outCount)))
  print("Time elapsed: %s seconds" % (time.time() - start))

  if (num_points <= 1000000):
    fig, ax = plt.subplots(1)
    ax.set_aspect('equal')
    ax.scatter(x_inside, y_inside, c='b', alpha=0.8, edgecolors=None) # points inside the circle will be colored blue 'b'
    ax.scatter(x_outside, y_outside, c='r', alpha=0.8, edgecolors=None) # points outside the circle will be colored red 'r'
```

```python
var2_plot(100)
var2_plot(2500)
var2_plot(10000)
var2_plot(1000000)
```

Here is the execution time of variant 2 as input size increases:

```python
var2_timeplot = list(zip(*var2_time))
plt.xlabel('Time (in seconds)')
plt.ylabel('Input Size')
plt.plot(var2_timeplot[1], var2_timeplot[0])
```

The second algorithm has the time complexity of O(n) since the input is also the range of the function. The function also only made use of a single loop to approximate the value of pi, hence, its time complexity O(n). 

### Variant 3 algorithm

The third algorithm computes the approximation of pi by considering the real values of x and y up to the given input size. The next cell will show us the code for the third algorithm.

```python
def var3(r):
  start = time.time()
  inCount = 0

  for i in range(1, r+1):
    x = i/r
    xsquare = x*x
    for j in range (1, r+1):
      y = j/r
      if (xsquare + y*y <= 1):
        inCount+=1
  print("--------------------Var3--------------------")
  size = r*r
  print("input size: ", size)
  print("pi value: ", (4*inCount/(r*r)))
  end = time.time() - start
  print("Time elapsed: %s seconds" % end)
  return size, end
  
```

Let us see the approximation of pi if we passed the same input size as the first two algorithms.

```python
var3_time.append(var3(100))
var3_time.append(var3(250))
var3_time.append(var3(500))
var3_time.append(var3(750))
var3_time.append(var3(1000))
var3_time.append(var3(2500))
var3_time.append(var3(5000))
var3_time.append(var3(7500))
var3_time.append(var3(10000))
```

The function below is the code for plotting the data points of the third algorithm

```python
def var3_plot(r):
  start = time.time()
  x_inside = []
  y_inside = []
  x_outside = []
  y_outside = []
  inCount = 0

  for i in range(1, r+1):
    x = i/r
    xsquare = x*x
    for j in range (1, r+1):
      y = j/r
      if (xsquare + y*y <= 1):
        x_inside.append(x)
        y_inside.append(y)
        inCount+=1
      else:
        x_outside.append(x)
        y_outside.append(y)
  print("--------------------Var3 w/ Visualization--------------------")
  print("input size: ", r*r)
  print("pi value: ", (4*inCount/(r*r)))
  print("Time elapsed: %s seconds" % (time.time() - start))

  if (r <= 1000000):
    fig, ax = plt.subplots(1)
    ax.set_aspect('equal')
    ax.scatter(x_inside, y_inside, c='b', alpha=0.8, edgecolors=None) # points inside the circle will be colored blue 'b'
    ax.scatter(x_outside, y_outside, c='r', alpha=0.8, edgecolors=None) # points outside the circle will be colored red 'r'
```

```python
var3_plot(10)
var3_plot(50)
var3_plot(100)
var3_plot(1000)
```

Here is the execution time of variant 3 as input size increases:

```python
var3_timeplot = list(zip(*var3_time))
plt.xlabel('Time (in seconds)')
plt.ylabel('Input Size')
plt.plot(var3_timeplot[1], var3_timeplot[0])
```

Similar to the first algorithm, the third algorithm has the same time complexity of O(r^2) rather than O(n^2) since we used a proxy for the size of the input which is r. The function also utilized nested loops with the range of r which resulted to a O(r^2) time complexity.
